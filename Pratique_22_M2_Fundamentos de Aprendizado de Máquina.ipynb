# **Exercícios**

## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
import numpy as np
import pandas as pd
import seaborn as sns
```

```
data = sns.load_dataset('penguins')
```

```
data.head()
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
colunas_numericas = data.select_dtypes(include=['float64', 'int64']).columns
data[colunas_numericas] = data[colunas_numericas].round(1)
data[colunas_numericas] = data[colunas_numericas].fillna(data[colunas_numericas].mean())
```

```
data['sex'].fillna('-', inplace=True)
```

```
data.head()
```

```
pinguins_df = pd.DataFrame(data)
```

```
print(pinguins_df.isna().sum())
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
pinguins_df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].head()
```

```
media_bill_lenght = pinguins_df['bill_length_mm'].mean()
print(media_bill_lenght)

desvio_padrao_bill_lenght = pinguins_df['bill_length_mm'].std()
print(desvio_padrao_bill_lenght)

pinguins_df['bill_length_mm_std'] = pinguins_df['bill_length_mm'].apply(lambda x: (x - media_bill_lenght) / desvio_padrao_bill_lenght)

```

```
media_bill_depth = pinguins_df['bill_depth_mm'].mean()
print(media_bill_depth)

desvio_padrao_bill_depth = pinguins_df['bill_depth_mm'].std()
print(desvio_padrao_bill_depth)

pinguins_df['bill_depth_mm_std'] = pinguins_df['bill_depth_mm'].apply(lambda y: (y - media_bill_depth) / desvio_padrao_bill_depth)
```

```
media_flipper_lenght = pinguins_df['flipper_length_mm'].mean()
print(media_flipper_lenght)

desvio_padrao_flipper_lenght = pinguins_df['flipper_length_mm'].std()
print(desvio_padrao_flipper_lenght)

pinguins_df['flipper_length_mm_std'] = pinguins_df['flipper_length_mm'].apply(lambda z: (z - media_flipper_lenght) / desvio_padrao_flipper_lenght)
```

```
media_body_mass = pinguins_df['body_mass_g'].mean()
print(media_body_mass)

desvio_padrao_body_mass = pinguins_df['body_mass_g'].std()
print(desvio_padrao_body_mass)

pinguins_df['body_mass_g_std'] = pinguins_df['body_mass_g'].apply(lambda w: (w - media_body_mass) / desvio_padrao_body_mass)
```

```
pinguins_df.head()
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
pinguins_df[['species', 'island', 'sex']].head()
```

```
pinguins_df['species_adelle_nom'] = pinguins_df['species'].apply(lambda species: 1 if species == 'Adelie' else 0)
pinguins_df['species_chinstrap_nom'] = pinguins_df['species'].apply(lambda species: 1 if species == 'Chinstrap' else 0)
pinguins_df['species_gentoo_nom'] = pinguins_df['species'].apply(lambda species: 1 if species == 'Gentoo' else 0)
```

```
pinguins_df['island_torgersen_nom'] = pinguins_df['island'].apply(lambda island: 1 if island == 'Torgersen' else 0)
pinguins_df['island_biscoe_nom'] = pinguins_df['island'].apply(lambda island: 1 if island == 'Biscoe' else 0)
pinguins_df['island_dream_nom'] = pinguins_df['island'].apply(lambda island: 1 if island == 'Dream' else 0)
```

```
pinguins_df['sex_male_nom'] = pinguins_df['sex'].apply(lambda sex: 1 if sex == 'Male' else 0)
pinguins_df['sex_female_nom'] = pinguins_df['sex'].apply(lambda sex: 1 if sex == 'Female' else 0)
```

```
pinguins_df.head()
```

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*".
