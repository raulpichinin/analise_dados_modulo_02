## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[['cut', 'price']].groupby('cut').agg('sum').reset_index()
diamonds

```

```
grafico = sns.barplot(data=diamonds, x='price', y='cut', palette='pastel')
grafico.set(title='Preço do Diamante por tipo de corte', xlabel='Tipo de Corte', ylabel='Preço');


```

**Insight do gráfico 1**


*   A diferença é pequena entre os tipos de corte Ideal e Premium.
*   Já do corte Good para Very Good a diferça é grande no valor.
*   O corte Fair é o mais barato de todos.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")
with sns.axes_style('whitegrid'):
  grafico = sns.lineplot(data=flights, x='month', y='passengers', hue='year', palette='muted')
  grafico.set(title='Passageiros por mês por ano entre 1949 e 1959 ', xlabel='Mês', ylabel='Passageiros')
  grafico.get_legend().set_title('Ano')
  grafico.legend(bbox_to_anchor=(1, 1), loc=2)
```

**Insight do gráfico 3**


*   Julho e Agosto são os meses que possuem a maior quantidade de passageiros isso para todos os anos. Provavelmente por conta das férias e estação do ano.
*   Aumento do numero de passageiros com o passar de cada ano.
*   Menor numero de passageiros nos meses de Fevereiro, Maio e Novembro.
