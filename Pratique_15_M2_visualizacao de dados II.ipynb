## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# ignorar avisos
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas
```

```
# gráfico do exercício 1
# para este exercício utilizarei um gráfico de distribuição - bloxplot*
with sns.axes_style('whitegrid'):

  diamonds = sns.boxplot(x=data["price"], y=data["cut"], palette="colorblind")
  diamonds.set(title="Distribuição de valores por corte de diamante", xlabel="Valores", ylabel="Tipo de Corte");
```

**Insight do gráfico 1**:

* Valores atípicos (outliers) por corte:
 - Ideal : A partir de 10500 aproximadamente
 - Premium: A partir de 14000 aprox.
 - Very Good: A partir de 12000 aprox.
 - Good: A partir de 11000 aprox.
 - Fair: A partir de 9900 aprox.
*   O preço comum entre todos tipos de corte está entre 2000 e 5000.
*   Premium é o corte com maior faixa de preços e Fair a menor faixa.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
diamonds = data[["price", "carat", "clarity"]]
diamonds.head()
```

```
# gráfico do exercício 2
# para este exercício utilizarei um gráfico de correlação - scatterplot*
with sns.axes_style('whitegrid'):

  diamonds = sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", palette="deep")
  diamonds.set(title='Correlação entre Preço do Diamante com seu Peso, agrupados por sua Transparêcia', xlabel='Peso', ylabel='Preço');
  diamonds.get_legend().set_title("Transparência");
```

**Insight do gráfico 2**:


*   Quanto mais pesado o diamante, maior seu preço.
*   Quanto mais transparente o diamante, mais raro e maior se preço.
*   Todas as claridades apresentam crescimento próximos, com exeção da I1.
*   I1 é a claridade mais rara entre as outras.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3
# para este exercício utilizarei um gráfico de distribuição - KDE (Kernel Density Estimator)*

with sns.axes_style('whitegrid'):

   cars = sns.histplot(data=data, x="weight", kde=True)
   cars.set(title='Distribuição contínua aproximada do peso dos carros', xlabel='PESO', ylabel='QUANTIDADE');
```

**Insight do gráfico 3**:


*   O peso médio dos carros é de aproximadamente 2300.
*   Há um crescimento no peso de 1600 até 2200 aproximadamente, logo depois ocorre uma queda contínua até aproximadamente 5250.
*   Não existe carros com peso maior que 5300.
